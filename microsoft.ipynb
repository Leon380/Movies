```python
q = """
SELECT 
      mb.movie_id,
      mb.primary_title,
      mr.averagerating,
      mr.numvotes,
      mb.genres
FROM movie_basics AS mb
JOIN movie_ratings AS mr
      ON mb.movie_id = mr.movie_id 
;
"""
df = pd.read_sql(q, conn)
df.head()
```
